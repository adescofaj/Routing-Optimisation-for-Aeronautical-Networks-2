### Converting into Cartesian Co-ordinates

<p style="font-family:Georgia;">The Cartesian function formula for three sets of numbers is:</p>
<p style="font-family:Georgia;">p<sub>x,a</sub> = (R<sub>E</sub> + L<sub>a</sub>) cos(θ<sub>a</sub>) cos(ϕ<sub>a</sub>)</p>
<p style="font-family:Georgia;">p<sub>y,a</sub> = (R<sub>E</sub> + L<sub>a</sub>) cos(θ<sub>a</sub>) sin(ϕ<sub>a</sub>)</p>
<p style="font-family:Georgia;">p<sub>z,a</sub> = (R<sub>E</sub> + L<sub>a</sub>) sin(θ<sub>a</sub>) 

```python
R = 6371000 
def erthCartes(links): 
    lon = pd_data._get_value(links,'Longitude') 
    lat = pd_data._get_value(links,'Latitude') 
    Lo = np.deg2rad(lon) 
    La =  np.deg2rad(lat)  
    Al = pd_data._get_value(links,'Altitude') * 0.304 
    X = (R + Al)* math.cos(La) * math.cos(Lo) 
    Y = (R + Al)* math.cos(La) * math.sin(Lo) 
    Z = (R + Al)* math.sin(La) 
    return(X,Y,Z)
```

### Function to find the distace between two points

The function begins by extracting the values of X, Y, and Z from the dataframe using the index A. It then calculates the Euclidean distance between the two points using the formula:
<p style="font-family:Georgia;">d = √((X<sub>GS1</sub> - X)<sup>2</sup> + (Y<sub>GS1</sub> - Y)<sup>2</sup> + (Z<sub>GS1</sub> - Z)<sup>2</sup>)
<p style="font-family:Georgia;">where d is the distance between the points (X<sub>GS1</sub>, Y<sub>GS1</sub>, Z<sub>GS1</sub>) and (X, Y, Z).

```python
def ode_edoBetweenPoints(links1,links2): 
    
    x1 = pd_data.loc[links1,'X'] 
    y1 = pd_data.loc[links1,'Y'] 
    z1 = pd_data.loc[links1,'Z'] 
 
    x2 = pd_data.loc[links2,'X'] 
    y2 = pd_data.loc[links2,'Y'] 
    z2 = pd_data.loc[links2,'Z'] 
     
    return math.sqrt((x1-x2)**2 + (y1-y2)**2 + (z1-z2)**2) 
```

The code defines a function "pastSem_gymenss" that takes a population of link permutations and a height value as inputs, and computes the best fitness score and corresponding "ode_edo" nodes for the population using a network packing algorithm and a transportation algorithm that considers both the transportation cost and the droppability of the packages.
```python
def pastSem_gymenss(aabd_Pop, height_0): 
    best_gymenss = 0 
    for publix in aabd_Pop: 
        ode_edos = netPack(publix, height_0)
        
         
        gymenss_sum = 0 
        for ode_edo in ode_edos: 
            tr = tr_ode_edo(ode_edo, height_0) 
            la = droppcy_ode_edo(ode_edo) 
            wsf = 0.4 * tr + 0.6 *( 1 / (la +1)) 
            gymenss_sum += wsf 
 
        if best_gymenss == 0 or best_gymenss < gymenss_sum: 
            best_gymenss = gymenss_sum 
            best_ode_edos = ode_edos 
             
    return (best_gymenss, best_ode_edos)  
```

The code implements the ordered crossover genetic operator to create a new population of individuals from a selected subset of the previous population by exchanging genetic material between two parents in a specific way.

```python
def ordered_wrappping(chosen_two): 
    l = len(chosen_two[0]) 
    progenee = [] 
    for i in range(0,len(chosen_two)-2,2): 
        p1 = chosen_two[i] 
        p2 = chosen_two[i+1] 
         
        s = round(random.uniform(l/2,l)) 
        o1 = [] 
        o2 = [] 
         
        for j in p2: 
            if j not in p1[s:]: 
                o1.append(j) 
        o1 = o1 + p1[s:]     
         
        for k in p1: 
            if k not in p2[s:]: 
                o2.append(k) 
        o2 = o2 + p2[s:]   
         
        progenee.append(o1) 
        progenee.append(o2) 
 
    return progenee 
```

 Performs greedy ordered wrapping crossover on a list of two parents. It computes the average gymenss of the parents, and generates new offspring by selecting a random split point and exchanging genetic material between the parents. If the gymenss of the resulting offspring is greater than or equal to the average gymenss of the parents, it is added to the next generation. This process is repeated until the desired number of offspring is generated.

```python
def greedy_ordered_wrappping(chosen_two, height_0): 
    Avg_par_gymenss = 0 
    for i in chosen_two: 
        fit,_ = pastSem_gymenss([i], height_0) 
        Avg_par_gymenss += fit 
     
    l = len(chosen_two) 
    progenee = [] 
    ol = len(progenee) 
    i = 0 
    count = 0 
 
    while ol != l: 
        count += 1 
 
        p1 = chosen_two[i] 
        p2 = chosen_two[i+1] 
         
        s = round(random.uniform(l/2,l)) 
        o1 = [] 
        o2 = [] 
         
        for j in p2: 
            if j not in p1[s:]: 
                o1.append(j) 
        o1 = o1 + p1[s:]     
         
        for k in p1: 
            if k not in p2[s:]: 
                o2.append(k) 
        o2 = o2 + p2[s:] 
         
        fit1,_ = pastSem_gymenss([o1], height_0) 
        if fit1 >= Avg_par_gymenss or count >= 5: 
            progenee.append(o1) 
             
        fit2,_ = pastSem_gymenss([o2], height_0) 
        if fit2 >= Avg_par_gymenss or count >= 5: 
            progenee.append(o2) 
         
        ol = len(progenee) 
        if i != (l-2): 
            i += 1 
        else: 
            i = 0 
 
    return progenee 
```

The fittest individuals from the population, modifies them using genetic operations, and then calculates their fitness score in order to optimise genetic sequences. The genetic sequence with the greatest fitness score is then returned after repeating this process for a certain number of epochs. Greedy selection, ordered wrapping, and mutation via a swapping approach are some of the genetic operations. The best fitness score obtained at each epoch is printed by the function.

```python
def first_optimization(index_of_epochs, fltts_links_list, aabd_Pop_size,height_0):
    aabd_Pop = init_juntt(fltts_links_list, aabd_Pop_size) 
 
    best_gymenss, best_ode_edo = rectness_func(aabd_Pop,height_0)
    print(f'Generation-0 best gymenss score is {best_gymenss}') 
     
    for i in range(index_of_epochs): 
        guardian = greedy_preference(aabd_Pop, height_0) 
         
        progenee = greedy_ordered_wrappping(guardian,height_0) 
         
        changinged_epochs = vectRepl_changingion(progenee, height_0)
 
        current_gymenss, current_ode_edo = rectness_func(changinged_epochs,height_0)
         
        if current_gymenss > best_gymenss: 
            best_gymenss = current_gymenss 
            best_ode_edo = current_ode_edo 
         
        print(f'Generation-{i+1} best gymenss score is {best_gymenss}')
    
    return (best_ode_edo) 
```

By using several genetic processes including selection, wrapping, and mutation, the code improves genetic sequences. The objective is to identify the sequence that has the highest fitness score, which is calculated using the "pastSem_gymenss" function. The function prints the best fitness score obtained for each epoch once the optimisation is completed for a predetermined number of epochs. Four different inputs are required by the function: a numerical value, a list of floating links, a population size parameter and a height parameter.

```python
def second_optimization(index_of_epochs, fltts_links_list, aabd_Pop_size, height_0):
    aabd_Pop = init_juntt(fltts_links_list, aabd_Pop_size)
    best_gymenss, best_ode_edo = pastSem_gymenss(aabd_Pop,height_0)
    print(f'Generation-0 best fitness score is {best_gymenss}') 
     
    for i in range(index_of_epochs): 
        guardian = greedy_preference(aabd_Pop, height_0) 
         
        progenee = greedy_ordered_wrappping(guardian,height_0) 
         
        changinged_epochs = vectRepl_changingion(progenee, height_0)
 
        current_gymenss, current_ode_edo = pastSem_gymenss(changinged_epochs, height_0)
         
        if current_gymenss > best_gymenss: 
            best_gymenss = current_gymenss 
            best_ode_edo = current_ode_edo 
         
        print(f'Generation-{i+1} best fitness score is {best_gymenss}')
    
    return (best_ode_edo)            
```

### Writing a function to convert it into JSON file

```python
def outASJSON(ode_edos): 
    final_json = [] 
    for ode_edo in ode_edos: 
        gs_links = ode_edo[-1] 
        pairs_dict = {'Source' : pd_data.loc[ode_edo[0],'Flight No.']} 
        ode_edo_path = [] 
        flow_low_ =  dispatch_flow(gs_links,ode_edo[0], ode_edo[1]) 
        for i in range(len(ode_edo)-1): 
            flow =  dispatch_flow(ode_edo[i],ode_edo[i+1], ode_edo[i-1]) 
            ode_edo_path.append((pd_data.loc[ode_edo[i+1],'Flight No.'],flow)) 
            if flow < flow_low_: 
                flow_low_ = flow
        pairs_dict['routing path'] = tuple(ode_edo_path) 
        pairs_dict['End-to-end transmission rate'] = flow_low_ 
        final_json.append(pairs_dict) 
    return final_json 
```
